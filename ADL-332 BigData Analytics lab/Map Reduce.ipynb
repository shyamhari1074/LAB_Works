```python
def mapper(lines):
    mapped_data=[]
    for line in lines:
        words=line.strip().split()
        for word in words:
            mapped_data.append((word,1))
    return mapped_data
def reducer(mapped_data):
    reduced_data={}
    for word,count in mapped_data:
        if word in reduced_data:
            reduced_data[word]+=count
        else:
            reduced_data[word]=count
    return reduced_data
n=int(input("Enter the number of lines"))
data=[]
for i in range(n):
    line=input("Enter your sentences:- ")
    data.append(line)
mapped=mapper(data)
result=reducer(mapped)
for i in result.items():
    print(i)
```
